# Even or Prime

The problem statement is to classify 1000 numbers into even or prime and put them into a dictionary consisting lists of even or prime digits. 

```python
import numpy as np
```

```python
num_list = np.random.randint(low=1,high=5000,size=1000)
```

```python
num_list
```

```python
even = list()

prime = list()

for num in num_list:
    divisor = 2

    rem = num % divisor

    if rem == 0:
        even.append(num)

    else:
        for divisor in range (2,num):
            rem = num%divisor
        
            if rem==0:
                break
            else:
                if divisor == num-1:
                    prime.append(num)
                    break
```

```python
D = {'even':even, 'prime': prime}
```
