```python
import numpy as np
```

```python
num_list = np.random.randint(low=99,high=9999,size=1000)
```

```python
num_list
```

```python
even = list()

prime = list()

for num in num_list:
    divisor = list(range(2,num))

    rem = num % 2

    if rem == 0:
        even.append(num)

    else:
        for divisor in range (3,num):
            rem = num%divisor
        
            if rem==0:
                break
            else:
                divisor = divisor + 1
                if divisor == num-1:
                    prime.append(num)
```
