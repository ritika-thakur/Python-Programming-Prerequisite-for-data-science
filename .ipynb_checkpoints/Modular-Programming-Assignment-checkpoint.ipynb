```python
def prime_or_not(x):
    
    for divisor in range (2,x):
        
        if x % divisor == 0:
            
            return 0,x
        
        if divisor == (x-1):
            
            return 1,x
```

```python
prime = []

for num in numbers:
    
    y,z = prime_or_not(num)
    
    if y == 1:
        
        prime.append(z)
```

# ASSIGNMENT: 

 Generate 1000 random 10 digit hexadecimal numbers, determine the highest digit in each hexadecimal number. 

 Group all the hexadecimal numbers according to even highest digits, odd highest digits, and prime highest digits.

 Determine the frequency of the hexadecimal numbers having even highest digits, odd highest digits, and prime highest digits.



```python
Hexa='0123456789ABCDEF'

odd_highest_digit = []

even_highest_digit = []

prime_highest_digit = []
```

```python
string_lengths = np.random.randint(low=5, high=11, size=1000)
```

```python
def generate_hexadecimal_numbers(string_lengths):
    
    num_list =[]

    for i in range (0,1000):
        
        hexadecimal_number = ''
        
        for j in range (0,string_lengths[i]):
            
            hexadecimal_number = hexadecimal_number + Hexa[np.random.randint(low=0, high=16, size=1)[0]]
        
        num_list.append(hexadecimal_number)  
        
    return num_list    
```

```python
def grouping_numbers(x):
    

    number_in_list = generate_hexadecimal_numbers(x)
    
    for i in number_in_list:
        
        finding_highest_digit(i)
    
    Grouped = dict()
    
    Grouped['Hexadecimal numbers with even highest digits'] = even_highest_digit

    Grouped['Hexadecimal numbers with odd highest digits'] = odd_highest_digit

    Grouped['Hexadecimal numbers with prime highest digits'] = prime_highest_digit
    
    
    return Grouped
 
```

```python

def finding_highest_digit(p):
    
    q=len(p)
    
    sort_digits = []
    
    for k in p:
        
        sort_digits.append(k)
    
    sorted_number = sorted(sort_digits)
    
    if sorted_number[q-1]=='F':
        
        odd_highest_digit.append(p)
            
    elif sorted_number[q-1]=='E':
        
        even_highest_digit.append(p)
            
    elif sorted_number[q-1]=='D':
        
        prime_highest_digit.append(p)
            
    elif sorted_number[q-1]=='C':
        
        even_highest_digit.append(p)
            
    elif sorted_number[q-1]=='B':
        
        prime_highest_digit.append(p)

    elif sorted_number[q-1]=='A':
        
        even_highest_digit.append(p)
        
    else:
        
        if int(sorted_number[q-1])%2 == 0:
            
            even_highest_digit.append(p)
        
        elif int(sorted_number[q-1])%2 == 1:
            
            odd_highest_digit.append(p)
        
        elif int(sorted_number[q-1])==1:
            
            odd_highest_digit.append(p)
        
        else:
            
            prime_highest_digit.append(p)
    
    return (even_highest_digit, odd_highest_digit, prime_highest_digit)
```

```python
def frequency(r,s,t):

    D = dict()
    
    D['frequency of hexadecimal numbers with even highest digits'] = len(r)

    D['frequency of hexadecimal numbers with odd highest digits'] = len(s)

    D['frequency of hexadecimal numbers with prime highest digits'] = len(t)
    
    return D
```
